```python
import os
import random
import io
import numpy as np
import scipy as sp
import pandas as pd
from collections import Counter
from os import listdir
from os.path import isfile, join, isdir
from random import shuffle
import fitz
import re
from load_utils import load_data_recursive_pdf
```

```python
root_dir=r'E:\planning_validator_data\raw_data\sorted_doc_types\image_based\new_download-15-2-22\for_labelling_azure'
```

```python
documents, targets = load_data_recursive_pdf(root_dir)
```

```python
# convert target classes to integers
from sklearn import preprocessing
targets_encoder = preprocessing.LabelEncoder().fit(targets)
y = targets_encoder.transform(targets)

#check that high tfidf score words make sense
from collections import Counter
target_counter = Counter(targets)

from vectorizer import TfIdfVectorizer
vectorizer = TfIdfVectorizer().fit(documents, y)
corpus = vectorizer._make_corpus(documents, targets)
X = vectorizer.transform(corpus)
words = vectorizer.get_feature_names() 
for i,(x,d) in enumerate(zip(X,targets_encoder.classes_)):
    ids = np.argsort(x.todense().A * -1,axis=1).reshape(-1)
    sorted_words = [words[id] for id in ids]
    print('%35s  %2d  [#%4d]  %s'%(d, i, target_counter[d], ' '.join(sorted_words[:10])))
    
# split train test
from sklearn.model_selection import train_test_split
documents_train, documents_test, targets_train, targets_test = train_test_split(documents, y, test_size=.3, random_state=42)
print()
print('#train:%d    #test:%d    #classes:%d'%(len(documents_train),len(documents_test), len(set(y))))
```

```python
%%time
from vectorizer import TfIdfVectorizer
from vectorizer import CompositeTextVectorizer
char_tfidf_vectorizer = TfIdfVectorizer(n_components=200, analyzer='char_wb', ngram_range=(3,7))
tfidf_vectorizer = TfIdfVectorizer(n_components=200, ngram_range=(2,3)) 
composite_vectorizer = CompositeTextVectorizer(vectorizers_list=[char_tfidf_vectorizer, tfidf_vectorizer], n_components=300)
vectorizers_list = [char_tfidf_vectorizer, tfidf_vectorizer, composite_vectorizer]
```
